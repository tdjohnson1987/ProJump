```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DavidarmGYRO=pd.read_csv("D:/kth sports tech-1/Pro jump project/BeachVolley-20250914T074621Z-1-001/BeachVolley/David Arm/Gyro-2025_04_04-11_32_45.csv")
DavidarmMAG=pd.read_csv("D:/kth sports tech-1/Pro jump project/BeachVolley-20250914T074621Z-1-001/BeachVolley/David Arm/Magn-2025_04_04-11_32_45.csv")
DavidarmACC = pd.read_csv("D:/kth sports tech-1/Pro jump project/BeachVolley-20250914T074621Z-1-001/BeachVolley/David Arm/Acc-2025_04_04-11_32_45.csv")
pd.set_option('display.max_columns', 200)
display(DavidarmACC)
```

```python
timestamp_last4 = DavidarmACC.tail(5)

# Remove last 3 rows from original
Davidarm_new = DavidarmACC.iloc[:-5]

display(Davidarm_new)
display(timestamp_last4)
```

```python
Davidarm_new = Davidarm_new.dropna(axis=1, how="all")
display(Davidarm_new)
```

```python
import datetime as dt
first_unix = int(Davidarm_new["Timestamp"].iloc[0])

# Convert to datetime (in seconds)
first_time = dt.datetime.fromtimestamp(first_unix)

print("Unix:", first_unix)
print("Converted:", first_time)
```

```python
Davidarm_new["Timestamp"] = pd.to_datetime(Davidarm_new["Timestamp"], unit="s")
display(Davidarm_new)
```

```python
Davidarm_new["Timestamp"] = pd.to_datetime(Davidarm_new["Timestamp"], unit="s").dt.strftime("%H:%M:%S")
display(Davidarm_new)
```

```python
Davidarm_new["Seconds"] = pd.to_datetime(Davidarm_new["Timestamp"], unit="s").dt.second
display(Davidarm_new)
```

```python
#second2_data = Davidarm_new[Davidarm_new["Seconds"] == 2][["AccX", "AccY", "AccZ"]]
#display(second2_data)

print(Davidarm_new.loc[Davidarm_new["Seconds"] == 2, ["AccX", "AccY", "AccZ","Seconds"]])

# indices = np.where(Davidarm_new["Seconds"] == 2)

# # Print corresponding AccX, AccY, AccZ values
# print(Davidarm_new.loc[indices, ["AccX", "AccY", "AccZ",]])
```

```python
plt.figure(figsize=(12,6))
plt.plot(Davidarm_new["Seconds"], Davidarm_new["AccX"], label="AccX")

plt.xlabel("Seconds")
plt.ylabel("AccX")
plt.title("AccX vs Seconds")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(Davidarm_new["Seconds"], Davidarm_new["AccY"], label="AccY")

plt.xlabel("Seconds")
plt.ylabel("AccY")
plt.title("AccY vs Seconds")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(Davidarm_new["Seconds"], Davidarm_new["AccX"], label="AccX")
plt.plot(Davidarm_new["Seconds"], Davidarm_new["AccY"], label="AccY")
plt.plot(Davidarm_new["Seconds"], Davidarm_new["AccZ"], label="AccZ")

plt.xlabel("Seconds")
plt.ylabel("Acceleration")
plt.title("Acceleration vs Seconds")

plt.legend()  
plt.show()
```
