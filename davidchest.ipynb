```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

DavidchestACC = pd.read_csv("D:\kth sports tech-1\Pro jump project\BeachVolley-20250914T074621Z-1-001\BeachVolley\David Chest\Acc-2025_04_04-11_32_45.csv")
pd.set_option('display.max_columns', 200)
display(DavidchestACC)
```

```python
timestamp_last4 = DavidchestACC.tail(4)

# Remove last 3 rows from original
DavidchestACC_new = DavidchestACC.iloc[:-4]

display(DavidchestACC_new)
display(timestamp_last4)
```

```python
DavidchestACC_new = DavidchestACC_new.dropna(axis=1, how="all")
display(DavidchestACC_new)
```

```python
import datetime as dt
first_unix = int(DavidchestACC_new["Timestamp"].iloc[0])

# Convert to datetime (in seconds)
first_time = dt.datetime.fromtimestamp(first_unix)

print("Unix:", first_unix)
print("Converted:", first_time)
```

```python
DavidchestACC_new["Timestamp"] = pd.to_datetime(DavidchestACC_new["Timestamp"], unit="s")
display(DavidchestACC_new)
```

```python
DavidchestACC_new["Seconds"] = pd.to_datetime(DavidchestACC_new["Timestamp"], unit="s").dt.second
display(DavidchestACC_new)
```

```python
#second2_data = Davidarm_new[Davidarm_new["Seconds"] == 2][["AccX", "AccY", "AccZ"]]
#display(second2_data)

print(DavidchestACC_new.loc[DavidchestACC_new["Seconds"] == 31, ["AccX", "AccY", "AccZ","Seconds"]])

# indices = np.where(Davidarm_new["Seconds"] == 2)

# # Print corresponding AccX, AccY, AccZ values
# print(Davidarm_new.loc[indices, ["AccX", "AccY", "AccZ",]])
```

```python
DavidchestACC_new.insert(0, "S.No", range(1, len(DavidchestACC_new) + 1))
display(DavidchestACC_new)
```

```python


subset = DavidchestACC_new.head(500)

fig, axs = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

axs[0].plot(subset["S.No"], subset["AccX"], color="b", linewidth=0.8)
axs[0].set_ylabel("AccX")

axs[1].plot(subset["S.No"], subset["AccY"], color="g", linewidth=0.8)
axs[1].set_ylabel("AccY")

axs[2].plot(subset["S.No"], subset["AccZ"], color="r", linewidth=0.8)
axs[2].set_ylabel("AccZ")
axs[2].set_xlabel("Sample Number (S.No)")

plt.suptitle("Acceleration vs Sample Number (First 500 Samples)")
plt.grid(True)
plt.show()
```

```python
# Take only the first 500 samples
subset = DavidchestACC_new.head(500)

plt.figure(figsize=(12,12))

plt.plot(subset["S.No"], subset["AccX"], label="AccX", linewidth=0.8)
plt.plot(subset["S.No"], subset["AccY"], label="AccY", linewidth=0.8)
plt.plot(subset["S.No"], subset["AccZ"], label="AccZ", linewidth=0.8)

plt.xlabel("Sample Number (S.No)")
plt.ylabel("Acceleration")
plt.title("AccX, AccY, AccZ vs Sample Number (First 500 Samples)")
plt.legend()
plt.grid(True)
plt.show()
```
